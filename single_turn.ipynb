```python
## import NOVA API Keys and necessary libraries 
from nova_key import NOVA_API_KEY
import os
from openai import OpenAI 

NOVA_MODEL_ID = 'nova-lite-v2'
```

```python
## Instantiate the client for API calls
client = OpenAI(
    api_key= NOVA_API_KEY,
    base_url="https://api.nova.amazon.com/v1",
    default_headers={"Accept-Encoding": "gzip , deflate, br"}
)
```

## Single Turn Interaction

```python
## Make the first single turn call to say hello 
response = client.chat.completions.create(
    model=NOVA_MODEL_ID,
    messages=[{
            "role": "system",
            "content": "You are a helpful assistant"
        },
        {
            "role": "user",
            "content": "Hello"
        },
    ],
    stream=False
)

print(response.choices[0].message.content)
```

```python
## Ask a question 
response = client.chat.completions.create(
    model="nova-pro-v1",
    messages=[{
            "role": "system",
            "content": "You are a helpful GenAI expert."
        },
        {
            "role": "user",
            "content": "What are the benefits of Agentic AI in the Generative AI domain. Keep the response under 200 words."
        },
    ],
    stream=False
)

print(response.choices[0].message.content)
```
