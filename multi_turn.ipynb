```python
## import NOVA API Keys and necessary libraries 
from nova_key import NOVA_API_KEY
from openai import OpenAI
import json
```

```python
NOVA_MODEL_ID = 'nova-lite-v2'
```

```python
## Instantiate the client for API calls
client = OpenAI(
    api_key= NOVA_API_KEY,
    base_url="https://api.nova.amazon.com/v1",
    default_headers={"Accept-Encoding": "gzip , deflate, br"}
)
```

## Multi Turn Interaction

```python
## Make the first single turn call to say ask for Penni Pasta recipe for 1 person
messages = [{
            "role": "system",
            "content": "You are a helpful chef"
        },
        {
            "role": "user",
            "content": "How should I make Penni pasta 1 for person"
        },
    ]

response = client.chat.completions.create(
    model=NOVA_MODEL_ID,
    messages=messages,
    stream=False
)

```

```python
# Model responds back with recipe
print(response.choices[0].message.content)
```

```python
## Append the response to messages
messages.append({
    "role": "assistant",
    "content": response.choices[0].message.content
})
```

```python
## Append a follow up question to the ongoing chat
messages.append({
    "role": "user",
    "content": "Can you change this recipe for four people?"
})
print(json.dumps(messages, indent=2))
```

```python
response = client.chat.completions.create(
    model=NOVA_MODEL_ID,
    messages=messages,
    stream=False
)
## Model responds back with updated recipe for 4 people
print(response.choices[0].message.content)
```
